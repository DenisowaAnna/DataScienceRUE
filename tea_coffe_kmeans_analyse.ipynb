```python
def prediction(): 
    data = pd.read_excel('Группы 22 _ 1.xlsx')
    new_df = {'Unnamed: 0': input('Имя: '),
        'Пол': input('Пол: '),
        'Цвет глаз': input('Цвет глаз: '), 
        'Высшая школа': input('Факультет: '),
        'Округ': input('Oкруг: '), 
        'Спорт': int(input('Спорт: ')), 
        'Во сколько встаете': int(input('Во сколько встаете: ')), 
        'Курение': int(input('Курение: ')),
        'К/ч': ''}
    data = data.append(new_df, ignore_index = True)
    filtred_data = data[data['Высшая школа'].notnull()]
    filtred_data['К/ч'] = filtred_data['К/ч'].apply(lambda x: 0 if x == 'к' else 1)
    d = {}
    for i, j in enumerate(pd.unique(filtred_data['Высшая школа'])):
        d[j] = i
    filtred_data['Высшая школа'] = filtred_data['Высшая школа'].apply(lambda x: d[x])
    filtred_data['Пол'] = filtred_data['Пол'].apply(lambda x: 0 if x == 'м' else 1)
    d_eye = {}
    for i, j in enumerate(pd.unique(filtred_data['Цвет глаз'])):
        d_eye[j] = i
    filtred_data['Цвет глаз'].fillna('Карий', inplace = True)
    filtred_data['Цвет глаз'] = filtred_data['Цвет глаз'].apply(lambda x: d_eye[x])
    d_dis = {}
    for i, j in enumerate(pd.unique(filtred_data['Округ'])):
        d_dis[j] = i
    filtred_data['Округ'] = filtred_data['Округ'].apply(lambda x: d_dis[x])
    filtred_data['Цвет глаз'] = filtred_data['Цвет глаз'] / np.linalg.norm(filtred_data['Цвет глаз'])
    filtred_data['Высшая школа'] = filtred_data['Высшая школа'] / np.linalg.norm(filtred_data['Высшая школа'])
    filtred_data['Округ'] = filtred_data['Округ'] / np.linalg.norm(filtred_data['Округ'])
    filtred_data['Во сколько встаете'] = filtred_data['Во сколько встаете'] / np.linalg.norm(filtred_data['Во сколько встаете'])
    i = len(filtred_data) - 1
    pred = []
    dist = []
    filtred_data = filtred_data.reset_index(drop=True)
    for j in range(0, 22):
        eye = (filtred_data.loc[i, 'Цвет глаз'] - filtred_data.loc[j, 'Цвет глаз'])**2
        sex = (filtred_data.loc[i, 'Пол'] - filtred_data.loc[j, 'Пол'])**2
        wakeup = (filtred_data.loc[i, 'Во сколько встаете'] - filtred_data.loc[j, 'Во сколько встаете'])**2
        smok = (filtred_data.loc[i, 'Курение'] - filtred_data.loc[j, 'Курение'])**2
        dist.append((math.sqrt(sum([eye, sex, wakeup, smok])), j))
    dist.sort()
    f = dist[0][1]
    s = dist[1][1]
    t = dist[2][1]
    ff = dist[3][1]
    tt = dist[4][1]
    ans = Counter([filtred_data.loc[f, 'К/ч'], filtred_data.loc[s, 'К/ч'], filtred_data.loc[t, 'К/ч'], filtred_data.loc[ff, 'К/ч'], filtred_data.loc[tt, 'К/ч']]).most_common()[0][0]
    return 'чай' if ans == 1 else 'кофе' 
```
